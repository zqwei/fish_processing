### Add other information from the meta tables

```python
df_meta = pd.read_csv('Data/metadata.csv', index_col=0)
```

```python
df_meta.head()
```

```python
df_cell_type = pd.read_excel('Data/zebrafish_single-cell_metadata.xlsx')
```

```python
df_cell_type.head()
```

```python
df_cell_type['Plate'] = df_cell_type['cDNA_plate'].str[6:8].astype(int)
```

```python
df_cell_type['isGlia'] = df_cell_type['Plate_name'].str.contains('Red')
```

```python
df_cell_type['isNeuron'] = df_cell_type['Plate_name'].str.contains('Green')
```

```python
df_cell_type.head()
```

```python
df_meta['isGlia'] = pd.Series(np.array([False]*len(df_meta)))
df_meta['isNeuron'] = pd.Series(np.array([False]*len(df_meta)))
df_meta['cell_type'] = pd.Series(np.array([False]*len(df_meta)))
for index, n_cell in df_cell_type.iterrows():
    row_index = df_meta['plate']==n_cell['Plate']
    df_meta.loc[row_index, 'isGlia'] = n_cell['isGlia']
    df_meta.loc[row_index, 'isNeuron'] = n_cell['isNeuron']
    cell_type_str = ''
    if n_cell['isGlia']:
        cell_type_str = cell_type_str + 'G'
    if n_cell['isNeuron']:
        cell_type_str = cell_type_str + 'N'
    df_meta.loc[row_index,'cell_type'] = cell_type_str
    
```

```python
df_meta.head()
```

```python
df_meta.to_csv('Data/metadata_new.csv')
```
